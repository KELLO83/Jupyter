### 데이터분석을 위한 함수들 Reshape Data

* pd.concat([df1,df2],axis=1) # 리스트로 묶어서 전달한다
* pd.concat([df1,df2],join='inner') # default = 'outer'  inner로할시 교집합만 출력

```python
df1 = pd.DataFrame([1,2,3,4],['A','B','C','D']) # 데이터 1 2 3 4 index A B C D
df2 = pd.DataFrame([5,6,7,8],['A','B','C','D']) # 데이터 5 6 7 8 index A B C D

df3 = pd.concat([df1,df2],axis=1)  #열방향으로 추가한다 
df4 = pd.concat([df1,df2],axis=0) # 행방향으로 추가한다
display(df3)
display(df4)
```

```python
df1 = pd.DataFrame([1,2,3,4],['A','B','C','D']) # 데이터 1 2 3 4 index A B C D
df2 = pd.DataFrame([5,6,7,8,np.nan],['A','B','C','D','E']) # 데이터 5 6 7 8 index A B C D
df3 = pd.concat([df1,df2],axis=1)
df4 = pd.concat([df1,df2],axis=1,join='inner') # JOIN ='outer' 가 기본이며 join = 'inner' 사용할시 Nan행이 날라간다
display(df3)
display(df4)
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
               'A': ['A0', 'A4', 'A2', 'A3'],
              'B': ['B0', 'B4', 'B2', 'B3']})
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3','K4'],
                       'C': ['C0', 'C1', 'C2', 'C3','C4'],
                       'D': ['D0', 'D1', 'D2', 'D3','D4']})

display(df1,df2)
```

### pd.merge 
* pd.merge(df1,df2,how="---",on="---") how 병합방향 on 기준점 merge는 리스트가아니다 

```python
df3 = pd.merge(df1,df2,how='inner',on='key') # key를 기준으로 교집합인것들만 출력
display(df3)
```

```python
df3 = pd.merge(df1,df2,how='right',on='key') # df2를기준으로 합친다 key값기준
df3
```

```python
df3 = pd.merge(df1,df2,how='left',on='key') # df1을기준으로 key값 
df3
```

```python
df3 = pd.merge(df1,df2,how='outer',on='key') #합집합
df3
```

### df.pivot

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
res = df1.pivot(index="도시",columns="연도",values="인구") #행이 도시 열 연도 값은 값이 인구를 기준으로 정렬한다
res
```

```python
res = df1.pivot(index="도시",columns="연도",values=['인구','지역']) # 행 도시 열 연도 값 인구랑 지역 순으로 재정렬
res
```
