```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t1=np.arange(0,5,0.1)
t2=np.arange(0,5,0.02)
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
plt.figure(1) #그림의 번호를 지정 다른 figure가 나올때까지 유지된다
plt.subplot(211) # 2행
plt.plot(t1,f(t1),'r--',marker="o")
plt.figure(2)
plt.subplot(212)
plt.plot(t2,f(t2),'c--',marker="o")
plt.show()
```
