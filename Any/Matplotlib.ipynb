```python
x=np.arange(0,3*np.pi,0.1) # 0부터 np.pi 까지 간격 0.1
print(x)
y1=np.sin(x) #x에있는 모든요소들을 sin으로 해서 반환한다
y2=np.cos(x) #x에있는 모든요소들을 cos으로 해서 list반환
plt.plot(y1,label="sin",color='c',lw=5) #label sin 
plt.plot(y2,label="cos",color='k',lw=3,ls="--") #label cos

plt.legend() #label 출력
plt.title("SIN COS Plot",size=20) #제목 달기 제목의 사이즈 20pt 지정
plt.xlabel("x축 Axis",size=15) #x축 라벨달기
plt.ylabel("y축 Axis",size=15) #y축 라벨달기

plt.show()
```
