```python
tips = sns.load_dataset('tips')
sns.set_theme(style="darkgrid")
sns.regplot(x="total_bill",y='tip',data=tips) #x축 지불금액 y축 tip data = tips 에서 로드 
plt.show() # plt.show를 해주어야 보여진다
```

```python
tips = sns.load_dataset('tips')

sns.set_theme(style="darkgrid")
sns.relplot(x="total_bill",y='tip',hue='smoker',data=tips) 
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.set_theme(style="darkgrid")
sns.relplot(x="total_bill",y="tip",hue='smoker',style='time',data=tips)
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.set_theme(style="darkgrid")
sns.relplot(x="sepal_length",y='petal_length',kind='line',data=iris)
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.set_theme(style="darkgrid")
sns.relplot(x='sepal_length',y='petal_length',hue='species',kind='line',data=iris)
plt.show()
```
